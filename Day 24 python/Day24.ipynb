```python
# Sample Data
rides = (
    (101, "Ravi", "Bangalore", 250, 4.5),
    (102, "Anjali", "Hyderabad", 180, 4.2),
    (103, "Kiran", "Bangalore", 300, 4.8),
    (104, "Meena", "Chennai", 220, 4.0),
    (105, "Arjun", "Bangalore", 150, 3.9)
)


```

```python
# 1. Count total rides
total_rides = len(rides)
print("Total Rides:", total_rides)
```

```python
# 2. Extract rides from Bangalore
print("\nRides from Bangalore:")
for ride in rides:
    if ride[2] == "Bangalore":
        print(ride)
```

```python
# 3. Total revenue from Bangalore
revenue = 0
for ride in rides:
    if ride[2] == "Bangalore":
        revenue += ride[3]

print("\nTotal Revenue from Bangalore:", revenue)

```

```python
# 4. Highest rated driver
highest_ride = rides[0]
for ride in rides:
    if ride[4] > highest_ride[4]:
        highest_ride = ride

print("\nHighest Rated Ride:", highest_ride)
```

```python
city_data = {}

for ride in rides:
    city = ride[2]
    rating = ride[4]

    if city not in city_data:
        city_data[city] = [rating]
    else:
        city_data[city].append(rating)

# average calculate
for city in city_data:
    avg = sum(city_data[city]) / len(city_data[city])
    city_data[city] = round(avg, 1)

print(city_data)
```
