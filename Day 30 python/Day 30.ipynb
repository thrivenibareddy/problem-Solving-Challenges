```python
s = "abcabcbb"

char_set = set()
left = 0
max_length = 0

for right in range(len(s)):
    while s[right] in char_set:
        char_set.remove(s[left])
        left += 1

    char_set.add(s[right])
    max_length = max(max_length, right - left + 1)

print(max_length)
```

```python
lst = [1, 2, 2, 3, 4, 3, 5]

result = []
seen = set()

for num in lst:
    if num not in seen:
        result.append(num)
        seen.add(num)

print(result)
```

```python
s = "programming"

freq = {}

for ch in s:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

print(freq)
```

```python
words = ["cat", "dog", "elephant", "bat", "tiger"]

grouped = {}

for word in words:
    length = len(word)

    if length not in grouped:
        grouped[length] = []

    grouped[length].append(word)

print(grouped)
```
