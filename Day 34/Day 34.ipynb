```python
#Remove Duplicates & Preserve Order
ids = [101, 102, 101, 103, 102, 104]

unique_ids = []
seen = set()

for id in ids:
    if id not in seen:
        unique_ids.append(id)
        seen.add(id)

print(unique_ids)
```

```python
#Frequency Counter (Real-time analytics)
products = ["apple", "banana", "apple", "orange", "banana", "apple"]

frequency = {}

for item in products:
    frequency[item] = frequency.get(item, 0) + 1

print(frequency)
```

```python
#Find Top Scorer
marks = {"A": 85, "B": 92, "C": 88}

top_scorer = max(marks, key=marks.get)

print(top_scorer)
```

```python
#Common Elements in Two Lists
skills1 = ["Python", "SQL", "Excel"]
skills2 = ["SQL", "Power BI", "Python"]

common = list(set(skills1).intersection(set(skills2)))

print(common)
```
